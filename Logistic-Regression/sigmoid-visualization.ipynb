```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Function to update plot based on w and b
def plot_sigmoid(w=1, b=0):
    z = np.linspace(-10, 10, 100)  # Input values
    y = sigmoid(w * z + b)  # Compute sigmoid
    
    plt.figure(figsize=(10, 5))
    plt.plot(z, y, label=f'Sigmoid with w={w}, b={b}', color='blue')
    
    # Mark the initial guess (w=0, b=0)
    initial_y = sigmoid(0 * z + 0)  # Sigmoid at initial w=0, b=0
    plt.plot(z, initial_y, linestyle="dashed", color="gray", label="Initial Guess (w=0, b=0)")
    
    # Reference lines
    plt.axvline(0, color='black', linestyle='--', linewidth=0.8)  # x = 0 reference
    plt.axhline(0.5, color='black', linestyle='--', linewidth=0.8)  # y = 0.5 reference
    
    # Labels and legend
    plt.title("Effect of Changing w and b on Sigmoid Function")
    plt.xlabel("z")
    plt.ylabel("Sigmoid(z)")
    plt.legend()
    plt.grid()
    plt.show()

# Interactive plot with sliders for w and b
interact(plot_sigmoid, w=widgets.FloatSlider(min=-5, max=5, step=0.5, value=1), 
                      b=widgets.FloatSlider(min=-5, max=5, step=0.5, value=0));
```
